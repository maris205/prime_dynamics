```python
import numpy as np

def test_hardy_littlewood_logistic():
    u_c = 1.5436890126920763
    N = 50_000_000  # 跑五千万步，确保不变测度极高精度收敛
    
    print(f"🚀 正在启动非线性动力学沙盘 (迭代步数: {N})...")
    
    # 1. 生成纯动力学轨道
    X = np.empty(N, dtype=np.float64)
    X[0] = 0.12345
    for _ in range(10000): # 烧掉瞬态
        X[0] = 1.0 - u_c * X[0]**2
        
    for i in range(1, N):
        X[i] = 1.0 - u_c * X[i-1]**2
        
    # 2. 映射到符号空间 (根据论文划分：x < 0 为 L, 即质数区)
    L = (X < 0)
    
    # 3. 捕捉不同间距的圆柱集 (用 Numpy 移位操作极其高效)
    # 孪生素数 (间距 2): L R L
    gap2 = L[:-2] & (~L[1:-1]) & L[2:]
    
    # 表兄弟素数 (间距 4): L R R R L
    gap4 = L[:-4] & (~L[1:-3]) & (~L[2:-2]) & (~L[3:-1]) & L[4:]
    
    # 性感素数 (间距 6): L R R R R R L
    gap6 = L[:-6] & (~L[1:-5]) & (~L[2:-4]) & (~L[3:-3]) & (~L[4:-2]) & (~L[5:-1]) & L[6:]
    
    # 4. 计算几何不变测度
    mu_2 = np.sum(gap2) / N
    mu_4 = np.sum(gap4) / N
    mu_6 = np.sum(gap6) / N
    
    print(f"\n=== 动力学几何测度 (抛物线分形概率) ===")
    print(f"mu_2 (孪生, 间距2) = {mu_2:.6f}")
    print(f"mu_4 (表亲, 间距4) = {mu_4:.6f}")
    print(f"mu_6 (性感, 间距6) = {mu_6:.6f}")
    
    print("\n=== 波利尼亚克 / Hardy-Littlewood 比例检验 ===")
    print(f"【数论真理】表亲/孪生 期望比例 = 1.000")
    print(f"【动力学实测】 mu_4 / mu_2 = {mu_4 / mu_2:.4f}")
    print("-" * 45)
    print(f"【数论真理】性感/孪生 期望比例 = 2.000 (因为6躲过了3的筛杀，密度应翻倍!)")
    print(f"【动力学实测】 mu_6 / mu_2 = {mu_6 / mu_2:.4f}")

if __name__ == "__main__":
    test_hardy_littlewood_logistic()
```
