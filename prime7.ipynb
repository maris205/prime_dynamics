```python
import matplotlib.pyplot as plt
import numpy as np

# 你的完美实验数据
k_values = np.arange(1, 15)
N_values = [9, 25, 49, 121, 169, 289, 361, 529, 841, 961, 1369, 1681, 1849, 2209]
density = [0, 0, 0.020833, 0, 0.005952, 0, 0, 0, 0, 0, 0, 0, 0, 0]

# 设置顶级学术期刊的极简绘图风格
plt.style.use('seaborn-v0_8-paper')
fig, ax = plt.subplots(figsize=(10, 6), dpi=150)

# 绘制火柴人图 (Stem Plot) 极其适合展示离散拓扑脉冲
markerline, stemlines, baseline = ax.stem(N_values, density, basefmt="k-", linefmt="#d62728", markerfmt="ro")
plt.setp(stemlines, 'linewidth', 2, 'alpha', 0.8)
plt.setp(markerline, 'markersize', 7)

# 添加视觉包络趋势线
ax.plot(N_values, density, color='#d62728', alpha=0.15, linewidth=4)

# 轴标签与标题
ax.set_title("Asymptotic Admissibility: Decay of Topological Defect Density $\\rho(N)$", fontsize=15, fontweight='bold', pad=15)
ax.set_xlabel("Physical Horizon $N$ (Squared Prime Boundary)", fontsize=13)
ax.set_ylabel("Defect Density $\\rho(N)$", fontsize=13)
ax.grid(True, linestyle='--', alpha=0.5)

# 致命绝杀：精准标注出大自然的“算术刺客”
ax.annotate('Topological Breakdown ($Q_3$)\nDriven by Gap (23-29)', 
            xy=(49, 0.020833), xytext=(200, 0.022),
            fontsize=11, color='#8c564b', fontweight='bold',
            arrowprops=dict(facecolor='#8c564b', shrink=0.05, width=1.5, headwidth=7))

ax.annotate('Topological Breakdown ($Q_5$)\nDriven by Large Gap (113-127)', 
            xy=(169, 0.005952), xytext=(500, 0.012),
            fontsize=11, color='#8c564b', fontweight='bold',
            arrowprops=dict(facecolor='#8c564b', shrink=0.05, width=1.5, headwidth=7))

# 标注宏观归零趋势
ax.annotate('Absolute Macroscopic Admissibility ($\\rho \\equiv 0$)', 
            xy=(1500, 0), xytext=(900, 0.005),
            fontsize=12, color='#2ca02c', fontweight='bold',
            arrowprops=dict(facecolor='#2ca02c', shrink=0.05, width=1.5, headwidth=7))

ax.set_ylim(-0.002, 0.025)
plt.tight_layout()
plt.show()
```
